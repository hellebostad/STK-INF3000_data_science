### k-week averages

```python
#Calculate the (k + 1)-week averages, defined by... Plot the 4- and 8-week averages and their deviations. What trends do you observe?
def k_week_averages(k,numbers):
    """Given k and a list of numbers, return k-week averages and deviations of list"""
    N = len(numbers)
    result = []
    deviations = []
    for i, number in enumerate(numbers): #enumerate is useful for obtaining an indexed list
        start = max(i-k, 0)
        end = min(i+1, N-1)
        nums = [numbers[l] for l in range(start,end)]
        result.append(sum(nums)/len(nums))
    for i, average in enumerate(result):
        deviations.append(numbers[i]-average)
    return result, deviations
```

```python
print len(messi)
print len(ronaldo)
```

```python
four_week_averages_messi, four_week_deviations_messi = k_week_averages(4, messi)
eight_week_averages_messi, eight_week_deviations_messi = k_week_averages(8, messi)
four_week_averages_ronaldo, four_week_deviations_ronaldo = k_week_averages(4, ronaldo)
eight_week_averages_ronaldo, eight_week_deviations_ronaldo = k_week_averages(8, ronaldo)
```

```python
plt.plot(month, messi, label = 'searches')
plt.plot(month, four_week_averages_messi, label = 'k = 4')
plt.plot(month, eight_week_averages_messi, label = 'k = 8')
plt.xlabel('Time, months')
plt.ylabel('Search interest (incl. averages)')
plt.legend(loc=0)
plt.gcf().autofmt_xdate()
plt.savefig('Messi_with_averages.png')
```

```python
#plot Messi vs Ronaldo 4-week averages
plt.plot(month, four_week_averages_messi, label = 'Messi')
plt.plot(month, four_week_averages_ronaldo, label = 'Ronaldo')
plt.title('Searches for Messi and Ronaldo')
plt.xlabel('Time, months')
plt.ylabel('Search interest, average')
plt.legend(loc=0)
plt.gcf().autofmt_xdate()
plt.savefig('Messi_vs_Ronaldo.png')
```

```python
#plot Messi vs Ronaldo 8-week averages
plt.plot(month, eight_week_averages_messi, label = 'Messi')
plt.plot(month, eight_week_averages_ronaldo, label = 'Ronaldo')
plt.title('Searches for Messi and Ronaldo')
plt.xlabel('Time, months')
plt.ylabel('Search interest, average')
plt.legend(loc=0)
plt.gcf().autofmt_xdate()
plt.savefig('Messi_vs_Ronaldo.png')
```

```python
# Plot deviations from averages
plt.plot(month, four_week_deviations_messi, label = 'Messi, k=4')
plt.plot(month, eight_week_deviations_messi, label = 'Messi, k=8')
plt.plot(month, four_week_deviations_ronaldo, label = 'Ronaldo, k=4')
plt.plot(month, eight_week_deviations_ronaldo, label = 'Ronaldo, k=8')
plt.xlabel('Time, month')
plt.ylabel('Deviation from average search interest')
plt.legend(loc=0)
plt.title('Deviation from average search interest')
plt.gcf().autofmt_xdate()
```

# Climate data - common search words over time

```python
climate_search = open('climate_search.csv')
```

```python
lines_to_print = 5
with open('climate_search.csv') as climate_search:
    for line in climate_search:
        print line
        lines_to_print -= 1
        if lines_to_print == 0:
            break
```

```python
#Create empty lists for search words to fill later:
month_climate = []
global_warming = []
climate = []
sustainability = []
climate_change = []
```

```python
def read_with_callback_climate(callback, start, end):
    with open('climate_search.csv') as climate_trends:
        for line in itertools.islice(climate_trends, start, end):
            callback(line)
```

```python
def fill_lists_climate(line):
    parts = line.split(",")
    month_climate.append(ymd(parts[0]))
    global_warming.append(int(parts[1]))
    climate.append(int(parts[2]))
    sustainability.append(int(parts[3]))
    climate_change.append(int(parts[4]))
read_with_callback_climate(fill_lists_climate, 1, None)
```

```python
# make a line plot
plt.plot(month_climate, global_warming, 'b', label = '"Global warming"')
plt.plot(month_climate, climate, 'g', label = '"Climate"')
plt.plot(month_climate, sustainability, 'y', label = '"Sustainability"')
plt.plot(month_climate, climate_change, 'm', label = '"Climate change"')
plt.title('Interest in climate seems to be declining')
plt.xlabel('Time, year and month')
plt.ylabel('Search interest')
plt.legend(loc=0, ncol=1)
plt.gcf().autofmt_xdate()
plt.savefig('Climate_keywords_overtime.png')
```

```python
four_week_averages_global_warming, four_week_deviations_global_warming = k_week_averages(4, global_warming)
eight_week_averages_global_warming, eight_week_deviations_global_warming = k_week_averages(8, global_warming)
four_week_averages_climate, four_week_deviations_climate = k_week_averages(4, climate)
eight_week_averages_climate, eight_week_deviations_climate = k_week_averages(8, climate)
```

```python
#plot global warming vs climate 8-week averages
plt.plot(month_climate, eight_week_averages_global_warming, label = '"Global Warming"')
plt.plot(month_climate, eight_week_averages_climate, label = '"Climate"')
plt.title('Searches for Global warming and climate')
plt.xlabel('Time, months')
plt.ylabel('Search interest, average')
plt.legend(loc=0)
plt.gcf().autofmt_xdate()
plt.savefig('globalwarming_vs_climate.png')
```

```python
# Plot deviations from averages
plt.plot(month_climate, four_week_deviations_global_warming, label = 'Global warming, k=4')
#plt.plot(month_climate, eight_week_deviations_global_warming, label = 'Global warming, k=8')
plt.plot(month_climate, four_week_deviations_climate, label = 'Climate, k=4')
#plt.plot(month_climate, eight_week_deviations_climate, label = 'Climate, k=8')
plt.xlabel('Time, month')
plt.ylabel('Deviation from average search interest')
plt.legend(loc=0)
plt.title('Deviation from average search interest')
plt.gcf().autofmt_xdate()
```
